Добавим индекс к каждому наблюдению

```python
df['id'] = np.arange(0, len(df))
df
```

```python
pivot1 = df.groupby('education').size().reset_index(name='ed_count')
pivot1
```

Посмотрим на распределение уровней образования

```python
ax = pivot1.plot.bar(
    x='education',
    y='ed_count',
    xlabel='level of education',
    ylabel='count',
    title='Распределение уровней образования'
)
fig1 = ax.get_figure()
fig1.savefig("../data/figures/education-count.png")
```

```python
pivot2 = df.groupby('age', dropna=True)['glucose'].mean()
pivot2.columns = ['age', 'glucose']
pivot2
```

Посмотрим на распределение уровня глюкозы по возрасту

```python
ax2 = pivot2.plot(
    x='age',
    y='glucose',
    xlabel='age',
    ylabel='glucose',
    title='Уровень глюкозы по возрасту'
)
fig2 = ax.get_figure()
fig2.savefig("../data/figures/glucose-age.png")
```
